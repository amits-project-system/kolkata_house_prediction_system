```python
def convert_price_to_crores(price_str):
    price_str = str(price_str).replace('₹', '').strip()
    if 'Cr' in price_str:
        return float(price_str.replace('Cr', '').strip())
    elif 'Lac' in price_str:
        return float(price_str.replace('Lac', '').strip()) / 100  # Convert Lac to Cr
    return np.nan # Handle any other cases, though based on value_counts, these are the main ones

data['Price'] = data['Price'].apply(convert_price_to_crores)
```

need to use pipeline to separate the numarical and catagorical data

```python
X=data.drop('Price(cr)',axis=1)
Y=data['Price(cr)']
```

```python
X.head(5)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=42)
```

##PREPROSSESING STEP:COLUMN TRANSFORMAR

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
```

```python
preprosser=ColumnTransformer(
    transformers=[
        ('categorical_data_trans', OneHotEncoder(handle_unknown='ignore', sparse_output=False), ['Location','Furnishing','House_Type'])
    ],
    remainder='passthrough'
)
```

```python
from sklearn.pipeline import Pipeline
```

##LinearRegression

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
model_pipeline=Pipeline(steps=[
    ('preprosser',preprosser),
    ('model',lr)
])
```

```python
from sklearn import set_config
set_config(display='diagram')
model_pipeline
```

```python
model_pipeline.fit(X_train,Y_train)
```

```python
Y_predict=model_pipeline.predict(X_test)
```

```python
from sklearn.metrics import r2_score
```

```python
r2_linear = r2_score(Y_test, Y_predict)
print("R² Score:", r2_linear)
```
